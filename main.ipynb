## Uji Reliabilitas Instrumen

Uji reliabilitas dilakukan menggunakan **Cronbach's Alpha** untuk mengukur konsistensi internal dari item-item pada instrumen.

### Catatan
- Nilai Alpha yang terlalu tinggi dapat mengindikasikan adanya item yang terlalu mirip atau berlebihan (redundansi), namun nilai 0.927 masih berada dalam batas yang ideal.
